# Simple series connection

```python
import schemdraw
import schemdraw.elements as elm

schemdraw.theme('monokai') # comment out to remove style

with schemdraw.Drawing():
    # 3 resistors with labels
    R1 = elm.Resistor().label('$R_1$\n10Ω')
    R2 = elm.Resistor().down().label('$R_2$\n50Ω', loc='bottom')
    R3 = elm.Resistor().left().label('$R_3$\n100Ω', loc='bottom')
    # source
    V = elm.SourceV().up().label(' V  \n20V')
    # current
    elm.LoopCurrent([R1, R2, R3, V], direction='cw').label('$I_T$')
```

# Parallel connection

```python
import schemdraw
import schemdraw.elements as elm

schemdraw.theme('monokai') # comment out to remove style

with schemdraw.Drawing() as d:
    # 3 resistors with labels
    L1 = elm.Line()
    elm.Dot()
    d.push()
    R1 = elm.Resistor().down().label('$R_1$\n100Ω', loc='bottom')
    d.pop()
    elm.Line()
    elm.Dot()
    d.push()
    R2 = elm.Resistor().down().label('$R_2$\n200Ω', loc='bottom')
    d.pop()
    elm.Line()
    elm.Dot()
    d.push()
    R3 = elm.Resistor().down().label('$R_3$\n600Ω', loc='bottom')
    elm.Dot()
    elm.Line().left()
    elm.Dot()
    elm.Line()
    elm.Dot()
    L6 = elm.Line()
    # source
    V = elm.SourceV().up().label(' V  \n60V')
    # current
    elm.LoopCurrent([L1, R1, L6, V],direction='cw').label('$I_T$')
```

# Series - Parallel connection

```python
import schemdraw
import schemdraw.elements as elm

schemdraw.theme('monokai')

with schemdraw.Drawing() as d:
    R1 = elm.Resistor().label('$R_1$\n25Ω')
    elm.Dot()
    # add R2 below
    d.push()
    R2 = elm.Resistor().down().label('$R_2$\n50Ω', loc='bottom')
    d.pop()
    # resume to the right
    R3 = elm.Resistor().label('$R_3$\n30Ω')
    R4 = elm.Resistor().down().label('$R_4$\n20Ω', loc='bottom')
    L1 = elm.Line().left()
    L2 = elm.Line()
    # source
    V = elm.SourceV().up().label(' V  \n20V')
    # current
    elm.LoopCurrent([R1, R2, L2, V],direction='cw').label('$I_T$')
```
